```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
corr_matrix = iris.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

>From the scatterplot matrix, we can observe that there are some clear differences between the three types of irises. For example, the scatterplots for petal length and petal width show a clear separation between Setosa and the other two types, while Versicolor and Virginica overlap with each other.
Additionally, the scatterplot for sepal length and sepal width doesn't show a clear separation between the three types, indicating that these two features may not be as useful in distinguishing between the iris types as the petal length and width.
The correlation coefficient matrix and heatmap show that there is a strong positive correlation between petal length and petal width, as well as between sepal length and petal length. There is also a weaker positive correlation between sepal length and sepal width. These correlations indicate that these features are not independent, and that they may be useful together in predicting the iris types.
Finally, the pie chart shows that the dataset is balanced, with each type of iris making up roughly one third of the total samples. This suggests that we don't need to worry about class imbalance when building a classification model.
